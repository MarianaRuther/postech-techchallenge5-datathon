## 4. Dataset original

O arquivo `BASE DE DADOS PEDE 2024 - DATATHON.xlsx` precisa estar em `data/raw/`.

````python
# Caminho do dataset
DATASET_PATH = PROJECT_ROOT / "data" / "raw" / "BASE DE DADOS PEDE 2024 - DATATHON.xlsx"

if DATASET_PATH.exists():
    tamanho_mb = DATASET_PATH.stat().st_size / (1024 * 1024)
    print(f"✅ Dataset encontrado")
    print(f"   Caminho: {DATASET_PATH}")
    print(f"   Tamanho: {tamanho_mb:.2f} MB")
else:
    print(f"❌ Dataset NÃO encontrado em:")
    print(f"   {DATASET_PATH}")
    print(f"\nMova o arquivo 'BASE DE DADOS PEDE 2024 - DATATHON.xlsx' pra dentro de data/raw/")
````

## 5. Prévia: inventário das abas do xlsx

Só uma espiadinha pra confirmar que a leitura funciona. A EDA completa vem na Etapa 2.

````python
if DATASET_PATH.exists():
    # Lista as abas do arquivo
    xl = pd.ExcelFile(DATASET_PATH)
    print(f"Abas encontradas: {len(xl.sheet_names)}\n")
    for i, nome in enumerate(xl.sheet_names, 1):
        print(f"  {i}. {nome}")
else:
    print("⚠️  Pule esta célula por enquanto — dataset ainda não está em data/raw/")
````

````python
# Prévia da primeira aba
if DATASET_PATH.exists():
    primeira_aba = xl.sheet_names[0]
    df_preview = pd.read_excel(DATASET_PATH, sheet_name=primeira_aba, nrows=5)
    print(f"Aba '{primeira_aba}' — shape prévio: {df_preview.shape}")
    print(f"\nColunas ({len(df_preview.columns)}):")
    for col in df_preview.columns:
        print(f"  - {col}")
    print("\nPrimeiras 5 linhas:")
    df_preview
````
